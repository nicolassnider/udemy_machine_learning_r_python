```python
#multiples graficos
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
x1=np.arange(0,5.0,0.1)
x2=np.arange(0,5.0,0.2)

plt.figure(1) # para crear figura conjunta

plt.subplot(211)# indica n° de filas y columnas
plt.plot(x1,f(x1),'ro',x2,f(x2),'k')

plt.subplot(212)
plt.plot(x2,f(x2),'g--')

plt.show()
```

```python
mu=100
sigma=20
x= mu + sigma * np.random.randn(10000)
```

```python
n,bins,patches=plt.hist(x,50,facecolor='g',alpha=0.6)
plt.xlabel('IQ')
plt.ylabel('Probabilidad')
plt.title(r"Hist de CI $N(\mu,\sigma)$")
#plt.axis ([20,180,0,0.025])
plt.text(120,0.015,r'$\mu = 100, \ \sigma = 20$')

plt.grid(True)
```

```python
plt.figure(figsize=(10,6),dpi=90)#pulgadas
plt.subplot(1,1,1)

x=np.arange(0,10*np.pi,0.01)
y=np.cos(x)
plt.plot(x,y,lw=2.0)

plt.annotate('Máximo local',xy=(4*np.pi,1),xytext=(15,1.5),arrowprops=dict(facecolor="black",shrink=0.05))
plt.ylim=(-2,2)
```

```python
#cambios de escala
from matplotlib.ticker import NullFormatter

mu=0.5
sd=0.3
y=mu + sd*np.random.randn(1000)
y=y[(y>0) & (y<1)]

x=np.arange(len(y))

plt.figure(1)

plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala lineal")
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale('log')
plt.title('escala log')

plt.subplot (2,2,3)
plt.plot(x,y-y.mean())
plt.yscale("symlog",lintreshy=0.01)
plt.title("E. Log simétrico")

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")
plt.title("E. logística")
plt.gca().yaxis.set_minor_formatter(NullFormatter())
plt.grid(True)


plt.subplots_adjust(top=0.92,bottom=0.08,left=0.1,right=0.95,hspace=0.35,wspace=0.35)
plt.show()
```
